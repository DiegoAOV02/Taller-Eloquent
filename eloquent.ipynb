```python
import pandas as pd
import tensorflow as tf
from tensorflow import keras
```

```python
# Carga de dataset
df = pd.read_csv('sensor_dataset.csv', header=None, names=['sensor_value', 'label'])
# Obtener caracteristicas de entrenamiento y etiquetas
x_train = df[['sensor_value']]
y_train = df[['label']]
```

```python
# Crear modelo

model = keras.Sequential([
    keras.layers.Dense(10, activation='relu', input_shape=(1,)),
    keras.layers.Dense(1)
])

model.compile(optimizer='adam', loss='mse')
```

```python
# Entrenar modelo

history = model.fit(x_train, y_train,epochs=1000)
```

```python
# Convertir modelo al formato tensorflow lite
converter = tf.lite.TFLiteConverter.from_keras_model(model)
tflite_model = converter.convert()

# Guardar modelo en el disco
open("moisture_sensor_model.tflite", "wb").write(tflite_model)
```
